```
import pandas as pd

# Creating citiesDataFrame

citiesDataFrame = pd.read_csv("cities.csv",quotechar='"',skipinitialspace=True)

citiesDataFrame
```

```
# Insert

# Example 1 with loc
citiesDataFrame.loc[len(citiesDataFrame.index)] = [41,22,34,"N",112,44,36,"W","Denizli","TR"]

citiesDataFrame.tail()
```

```
# Insert

# Example 2
data = {'LatD':50,'LatM':25,'LatS':40,'NS':"S",'LonD':125,'LonM':78,'LonS':45,'EW':"W",'City':"Izmir",'State':"TR"}
citiesDataFrame = citiesDataFrame.append(data,ignore_index=True)
citiesDataFrame.tail()
```

```
# Update

# Example 1
citiesDataFrame.loc[citiesDataFrame['City'] == "Youngstown",'State'] = "TR"
citiesDataFrame.head()
```

```
# Update

# Example 2
# citiesDataFrame[128].LatM = 22 => citiesDataFrame[128].LatM = 35
citiesDataFrame.loc[citiesDataFrame['LatM'] == 22,'LatM'] = 35
citiesDataFrame.tail()
```

```
# Delete

citiesDataFrame = citiesDataFrame.drop([127,129])

citiesDataFrame.tail()
```

```
# Reindexing primary key

citiesDataFrame = citiesDataFrame.reset_index(drop=True)
citiesDataFrame
```
